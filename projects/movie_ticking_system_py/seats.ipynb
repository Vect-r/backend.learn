```python
import customtkinter
import string
```

```python
seats={}
rows=string.ascii_uppercase[:12]
columns=list(str(i+1) for i in range(20))
buttons={}

for cr,row in enumerate(rows):
    for cc,col in enumerate(columns):
        seat=f"{row}{col.zfill(2)}"
        seats[seat]={'row':cr,'col':cc,'isAvailable':True,'isSelected':False}
```

```python
app=customtkinter.CTk()
app.title('Seat Choose')
app.geometry('800x800')
frame=customtkinter.CTkFrame(app)
frame.pack(pady=10,padx=10)
for label,seat in seats.items():
    buttons[label]=customtkinter.CTkButton(frame,text=label,fg_color="green",font=('Calibri',12),width=30,height=30,corner_radius=2)
    buttons[label].grid(row=seat['row'],column=seat['col'],padx=2,pady=2)

app.mainloop()
```
