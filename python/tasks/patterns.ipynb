```python
start,end=1,5
i=start
while(i<=end):
    j=i
    while(j>=start):
        print(j, end=" ")
        j-=1
    print("\n")
    i+=1
```

```python
# grid x grid
grid, start = 11, 1
pacePoint = round(grid / 2)
print(pacePoint)

# row
row = 1
while(row <= grid):
    col = start
    while (col <= grid):
        # print(f"col ", end=" ")

        if (row in range(start, pacePoint) and col == start) or (row in range(pacePoint + 1, grid + 1) and col == grid):
            print(f"* ", end="")
        elif((col or row) == pacePoint):
            print(f"* ", end="")
        else:
            print(f"  ", end="")

        col += 1
    print()
    row += 1
```
