```python
password=input("Enter Password: ")
isUpper=False
isLower=False
isDigit=False
isSpec=False
if len(password)>=8:
    for i in password:
        # print(i.islower())
        if i.isupper() and isUpper==False:
            isUpper=True
            continue
        if i.islower() and isLower==False:
            isLower=True
            continue
        if i.isdigit() and isDigit==False:
            isDigit=True
            continue
        if not i.isalnum() and isSpec==False:
            isSpec=True
        
    print(isUpper,isLower,isDigit,isSpec)
    if not isUpper: print("There are no Uppercase Chars")
    elif not isLower: print("There are no Lowercase Chars")
    elif not isDigit: print("There are no Digits")
    elif not isSpec: print("There are no Special Chars")
    else: print("Password Validated.")
else:
    print("Error. Password length should be at least 8 Chars Long")
```

```python
data = "this Is MY $1000"
result={'lower':0,
        'upper':0,
        'spec-char':0,
        'digits':0,
        'even-number':0,
        'odd-number':0}
for i in data:
    if i.isupper():
        result['upper']+=1
        continue
    if i.islower():
        result['lower']+=1
        continue
    if i.isdigit():
        result['digits']+=1
        if int(i)%2==0:
            result['even-number']+=1
        else:
            result['odd-number']+=1
        continue
    if not i.isalnum():
        result['spec-char']+=1
print(result)
```

```python
start,end=1,9
pacePoint=math.ceil(end/2)
f=1
for i in range(start,end+1):
    # print(i,f)
    if i>=pacePoint:
        print(" "*(i-1),"*"*f)
        f-=2
    else:
        print(" "*(end-i),"*"*f)
        f+=2
    
```
