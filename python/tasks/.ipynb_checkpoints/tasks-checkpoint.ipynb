```python
data = "this Is MY $1000"
result={'lower':0,
        'upper':0,
        'spec-char':0,
        'digits':0,
        'even-number':0,
        'odd-number':0}
for i in data:
    if i.isupper():
        result['upper']+=1
        continue
    if i.islower():
        result['lower']+=1
        continue
    if i.isdigit():
        result['digits']+=1
        if int(i)%2==0:
            result['even-number']+=1
        else:
            result['odd-number']+=1
        continue
    if not i.isalnum():
        result['spec-char']+=1
print(result)
```
